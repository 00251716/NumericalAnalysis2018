
# Apéndice

El objetivo principal de este apéndice es demostrar el teorema de Picard, que se utiliza para garantizar la existencia y unicidad de soluciones a problemas de valor inicial. Para facilitar la demostración, introducimos conceptos matemáticos importantes.

**Definición 1.**  Un *espacio métrico* es un conjunto $X$ con una función $d: X \times X \rightarrow [0, \infty ) $ tal que 

$\bullet \enspace d(x, y) $ si y solo si $y = x$

$\bullet \enspace d(x,y) = d(y, x) $ para todo $x$ y $y$ en $X$

$\bullet \enspace d(x, z) \leq d(x, y) + d(y, z)$ para todo $x$, $y$ y $z$ en $X$

La función $d$ se denomina métrica o función de distancia en $X$. Solemos escribir $(X, d)$ para denotar el conjunto $X$ con su métrica $d$.

El ejemplo clásico de espacio métrico es $\mathbb{R}$ con la función valor absoluto como métrica. Resulta que la definición anterior permite generalizar conceptos como convergencia y continuidad; eso es justo lo que haremos a continuación. Comenzamos recordando la definición de una sucesión.

**Definición 2.** Una sucesión $a$ en un conjunto $A$ es una función $a: \mathbb{N} \rightarrow A$. Usualmente escribimos $(a_{n})$ en lugar de $a$ y $a_{n}$ en lugar de $a(n)$.

**Definición 3.** Sea $(X, d)$ un espacio métrico. Una sucesión $(a_{n})$ en $X$ converge a $b \in X$ si para todo $\epsilon > 0$ existe un $N \in \mathbb{N}$ tal que para todo $n \geq N$ tenemos que $d(a_{n}, b) < \epsilon$. Entonces $b$ se denomina el límite de $(a_{n})$ y escribimos $\lim_{n \to \infty} a_{n} = b$ o $a_{n} \rightarrow b$ 

**Definición 4.** Una sucesión $(a_{n})$ en un espacio métrico $(X, d)$ es una sucesión de Cauchy si para todo $\epsilon > 0$ existe un $N \in \mathbb{N}$ tal que para cualesquiera $m, n \geq N$ tenemos que $d(a_{m}, a_{n}) < \epsilon$

**Definición 5.** Un espacio métrico $X$ es completo si toda sucesión de Cauchy en $X$ converge en $X$.

Ahora que hemos generalizado nociones de convergencia, también podemos generalizar la noción de continuidad.

**Definición 6.** Sean $(X, d_{X})$ y $(Y, d_{Y})$ espacios métricos. Una función $f: X \rightarrow Y$ es continua en $a\in X$ si tenemos que para todo $\epsilon > 0$ existe un $\delta > 0 $ tal que si $x\in X$ y $d_{X}(x, a) < \delta$ entonces $d_{Y}(f(x), f(a)) < \epsilon $.



**Definición 7.** Sean $(X, d_{X})$ y $(Y, d_{Y})$ espacios métricos. Una función $f: X \rightarrow Y$ es una contracción si existe un $\lambda \in [0, 1)$ tal que para todos los $x_{1}$ y $x_{2}$ tenemos que

$$ d_{Y} ( f(x_{1}), f(x_{2}) ) \leq \lambda d_{X}(x_{1}, x_{2}) $$ 

Después de tantas definiciones, al fin podemos demostrar un teorema importante. El siguiente resultado fue demostrado por Stefan Banach en 1922 y se suele conocer como teorema de punto de fijo de Banach o principio de contracción.

**Teorema 1.** Sea $(X, d)$ un espacio métrico completo no vacío y $f:X \rightarrow X$ una contracción. Entonces $f$ tiene un único punto fijo en $X$, es decir, un $p \in X$ tal que $f(p) = p$.
